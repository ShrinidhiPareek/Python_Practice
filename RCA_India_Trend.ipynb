```python
RCA=pd.read_csv("RCA_INDIA_TOP_FIVE.csv",names=column)
```

```python
RCA.drop(0,axis=0,inplace=True)
```

```python
RCA.dropna(axis=1,inplace=True)
```

```python
RCA.columns=RCA.columns.str.replace("_Index_Value","")#we needed to change them to year so now we will be able to change them to date time
#and can then individually plot them 
```

```python
RCA
```

```python
RCA.drop(['Product_Code','Product_Label'],axis=1,inplace=True)
```

```python
RCA_transposed=RCA.transpose()
```

```python
RCA_transposed.index=pd.to_datetime(RCA_transposed.index)
```

```python
RCA_transposed
```

```python
Engines_motors=pd.Series(RCA_transposed[1],name='engines').astype(float)
```

```python
names=['Engines','Telecom','Motor_vehicles','Vehicle_accesories','Ships']
```

```python
for i in range(0,5):
    names[i]=pd.Series(RCA_transposed[i+1],name=names[i]).astype(float)
    
```

```python
Engines=names[0]
Telecom=names[1]
Motor_vehicles=names[2]
Vehicle_accesories=names[3]
Ships=names[4]
```

```python
import matplotlib.pyplot as plt
```

```python
fig,(engines,telecom,motors,parts,ships)=plt.subplots(5,1,figsize=(8,20),layout='constrained')
fig.suptitle("India RCA Trends for Section 7 of SITC(Machinery and transport equipment)",fontsize=12)

#Creating plot for Engines&motors locomotie nes
engines.plot(Engines.index,Engines,marker='o',markersize='3',color='darkblue')
engines.set(title="Engines & motors, non-electric; parts, n.e.s.", ylabel="RCA Values", xlabel="Years",ylim=(0.0,2.0))

#Creating plot for Telecommunication equipment, n.e.s.; & parts
telecom.plot(Telecom.index,Telecom,marker='o',markersize='3',color='darkblue')
telecom.set(title=" Telecommunication equipment, n.e.s.; & parts",ylabel="RCA Values", xlabel="Years", ylim=(0.0,2.0))

# Creating plot for Motor vehicles for the transport of persons
motors.plot(Motor_vehicles.index,Motor_vehicles,marker='o',markersize='3',color='darkblue')
motors.set(title="Motor vehicles for the transport of persons",ylabel="RCA Values",xlabel="Years",ylim=(0.0,2.0))

# Creating plot for Parts & accessories of vehicles of 722, 781, 7..
parts.plot(Vehicle_accesories.index,Vehicle_accesories,marker='o',markersize='3',color='darkblue')
parts.set(title="Parts & accessories of vehicles of 722, 781, 782, 783",ylabel="RCA Values",xlabel="Years",ylim=(0.0,2.0))

# Creating plot for Ships, boats & floating structures	
ships.plot(Ships.index,Ships,marker='o',markersize='3',color='darkblue')
ships.set(title="Ships, boats & floating structures",ylabel="RCA Values",xlabel="Years",ylim=(0.0,3.0))
#Now since the xlabel and title of  subplots are overlapping we can correct this using
#fig.tight_layout()# why I did not use tight_layout is because it doesnot consider figsuptitle only consideres subplot where as layout attribute
#1in subplots does that'''
plt.savefig("India RCA Trends for Section 7 of SITC(Machinery and transport equipment)1.jpg")
plt.show()
```
